Now is the time to see the under-over performances

```python
#tempdf = data[["2018-2019 Math Scale Score", "2018-2019  VAM Target", "ALEKS", "Grade"]]
data["Performance"] = data["2018-2019 Math Scale Score"]- data["2018-2019  VAM Target Math"]
#tempdf = tempdf.dropna()

plt.figure(figsize=(10,7))
plt.title("Performance vs ALEKS", size=30)
plt.ylabel("ALEKS", size=20)
plt.xlabel("Performance", size=20)
sns.scatterplot(y="Performance", x="ALEKS", data=data)
plt.show()
```

It is hard to say there exists a correlation between students performances (how above or below of target they scored) below some ALEKS values. Let us examine this.

```python
diff = []
aleks_diff = []
for i in range(10,int(max(data.ALEKS)+1)):
    if data[data["ALEKS"]==i]["ALEKS"].count()!=0:
        
        d = data[(data["ALEKS"]<i) & (data["Performance"]<0)].shape[0] - data[(data["ALEKS"]<i) & (data["Performance"]>0)].shape[0]
        tot = data[(data["ALEKS"]<i) & (data["Performance"]<0)].shape[0] + data[(data["ALEKS"]<i) & (data["Performance"]>0)].shape[0]
        perc = d/tot*100
        diff.append(perc)
        aleks_diff.append(i)
```

```python
diff_Series = pd.Series(diff)
aleks_diff_Series = pd.Series(aleks_diff)
```

```python
plt.figure(figsize=(10,7))
plt.title("Percent Difference Under-Over Performed vs ALEKS", size=30)
plt.ylabel("Percent Difference ", size=20)
plt.xlabel("ALEKS", size=20)
sns.scatterplot(x=aleks_diff, y=diff_Series)
```
